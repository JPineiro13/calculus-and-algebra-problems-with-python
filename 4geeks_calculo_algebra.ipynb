```python
#4Geeks cálculo y álgebra

#Ej 1

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


x = np.linspace(0, 10, 16)
y = x * 2.5

plt.plot(x,y)
plt.xlabel = ('Tiempo en segundos')
plt.ylabel = ('Metros recorridos')
plt.title('Gráfico distancia-tiempo')
plt.show()
```

```python
df = pd.DataFrame({'Tiempo en segundos': x, 'Metros recorridos': y})
print(df)
```

```python
#Ej 2

#Tenemos que plotear un gráfico cuyos valores son un poco más complejos de obtener

v0 = 0
t = 10
a = v0 / t

def funcion_distancia(v0, a, t):
    return v0 * t + 0.5 * a * t**2

valores_tiempo = np.linspace(0, 10, 16)
valores_distancia = funcion_distancia(v0, a, valores_tiempo)
df = pd.DataFrame({'Tiempo en segundos': valores_tiempo, 'Metros recorridos': valores_distancia})
print(df)
```

```python
#Ej 3

#Muy parecido. Definir una función cuadrática y plotearla

def costes(x):
  return 0.1 * (x)**2 - 9*x + 4500

x = np.linspace(0,100)
plt.plot(x, costes(x))
```

```python
#Ej 4

#Sumar dos matrices

matriz_1 = np.array([[1, 2], [3, 4]])


matriz_2 = np.array([[4, 5], [6, 7]])

print(np.add(matriz_1, matriz_2))
```

```python
#Ej 5

#Sumar dos listas

ls_a = [2, 5, 4, 7, 3]
ls_b = [1, 4, 6, 9, 10]

from operator import add

suma_ls_a_b = [sum(i) for i in zip(ls_a, ls_b)]

print ("Resultant list is : " + str(suma_ls_a_b))
```
