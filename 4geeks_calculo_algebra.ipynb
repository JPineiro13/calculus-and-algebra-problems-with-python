```python
#Ej 2

#Tenemos que plotear un gráfico cuyos valores son un poco más complejos de obtener

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


D = 25
t_total = 10
a = D / 50


time_values = np.linspace(0, t_total, 100)

distance = 0.5 * a * time_values**2
speed = a * time_values

# Creo la tabla
data = {'Tiempo': time_values, 'Velocidad': speed, 'Distancia': distance}
df = pd.DataFrame(data)


print(df.head(10))

# Tenemos este conjunto de líneas con las que especificar bastantes parámetros del gráfico como tamaño, leyendas y etiquetas
plt.figure(figsize=(8, 5))
plt.plot(time_values, distance, label="Distancia en m")
plt.plot(time_values, speed, label="Velocidad en m/s", linestyle='--')
plt.title("Distancia y velocidad a lo largo del tiempo")
plt.xlabel("Time (s)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Ej 7 (?) Bonus

#Me estoy inventando este ejercicio para practicar. La persona ha conseguido incrementar gradualmente su velocidad,
# no obstante, parece que se ha detenido en seco, como si se hubiera chocado contra una puerta de cristal muy limpia
#En este ejercicio trataré de reflejar un recorrido más realista

#Constantes
distancia_total = 25
tiempo_total = 10
metros_recorridos_hasta_decelerar = 20
distancia_deceleracion = distancia_total - metros_recorridos_hasta_decelerar  # Podría haber escrito 5 directaemte
tiempo_aceleracion = 8  #Pongo 8s para mantener la proporción frente a la distancia, aunque no es necesario
tiempo_desaceleracion = tiempo_total - tiempo_aceleracion

#Primer tramo
aceleracion = metros_recorridos_hasta_decelerar / (0.5 * tiempo_aceleracion**2)  # Fórmula de distancia: D = 0.5 * a * t^2

#Segundo tramo. Decelerar
velocidad_a_20m = aceleracion * tiempo_aceleracion
desaceleracion = -(velocidad_a_20m**2) / (2 * distancia_deceleracion)

#Calculamos varios puntos para representarlos
tiempo_fase_aceleracion = np.linspace(0, tiempo_aceleracion, 20)
tiempo_fase_desaceleracion = np.linspace(tiempo_aceleracion, tiempo_total, 20)

#Distancia y velocidad durante el primer tramo
distancia_durante_aceleracion = 0.5 * aceleracion * tiempo_fase_aceleracion**2
velocidad_durante_aceleracion = aceleracion * tiempo_fase_aceleracion

#Distancia y velocidad durante el segundo
distancia_durante_desaceleracion = metros_recorridos_hasta_decelerar + (velocidad_a_20m * (tiempo_fase_desaceleracion - tiempo_aceleracion)) + (0.5 * desaceleracion * (tiempo_fase_desaceleracion - tiempo_aceleracion)**2)
velocidad_durante_desaceleracion = velocidad_a_20m + desaceleracion * (tiempo_fase_desaceleracion - tiempo_aceleracion)

#Aquí se unen ambos tramos con concatenate
tiempo_total_valores = np.concatenate((tiempo_fase_aceleracion, tiempo_fase_desaceleracion))
distancia_total_valores = np.concatenate((distancia_durante_aceleracion, distancia_durante_desaceleracion))
velocidad_total_valores = np.concatenate((velocidad_durante_aceleracion, velocidad_durante_desaceleracion))

# Crear el dataframe con la tabla de resultados
datos = {'Tiempo en s': tiempo_total_valores, 'Velocidad en m/s': velocidad_total_valores, 'Distancia en m': distancia_total_valores}
df = pd.DataFrame(datos)

print(df.head(10))


plt.figure(figsize=(10, 6))
plt.plot(tiempo_total_valores, distancia_total_valores, label="Distancia en m", color='blue')
plt.plot(tiempo_total_valores, velocidad_total_valores, label="Velocidad en m/s", linestyle='--', color='orange')
plt.title("Distancia y velocidad en escenario más realista")
plt.xlabel("Tiempo en s")
plt.legend()
plt.grid(True)
plt.show()

```
